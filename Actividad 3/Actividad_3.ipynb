## Ejercicio 3

Define una función llamada is_romano() dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano.

```python
import re
romanos = {'I': 1, 'V': 5, 'X': 10,
               'L': 50, 'C': 100, 'D': 500, 'M': 1000}

class Numero:      
    def __init__(self,numero):
        self.normal = numero
        self.romano = self.printRoman(numero)

    def imprime(self):
        print(self.normal,self.romano)

    def roman2int(self,romval):
        total = 0
        prev = 0
        for letra in romval[::-1]:
            valor = romanos[letra]
            total += valor if valor >= prev else -valor
            prev = valor

        return total
        

    def suma_romano(self,numeroRomano):
        numeroNormal=self.roman2int(numeroRomano)
        suma=numeroNormal+self.normal
        self.normal=suma
        self.romano=self.printRoman(suma)
               

    def printRoman(self,num):
        m = ["", "M", "MM", "MMM"]
        c = ["", "C", "CC", "CCC", "CD", "D",
         "DC", "DCC", "DCCC", "CM "]
        x = ["", "X", "XX", "XXX", "XL", "L",
         "LX", "LXX", "LXXX", "XC"]
        i = ["", "I", "II", "III", "IV", "V",
         "VI", "VII", "VIII", "IX"]
  
        # Converting to roman
        thousands = m[num // 1000]
        hundreds = c[(num % 1000) // 100]
        tens = x[(num % 100) // 10]
        ones = i[num % 10]
  
        ans = (thousands + hundreds +
           tens + ones)
  
        return ans

    def is_romano(self,caracteres):
        isRoman=bool(re.search(r"^M{0,3}(CM|CD|D?C{0,3})(XC|XL|L?X{0,3})(IX|IV|V?I{0,3})$",caracteres))
        res=False
        if isRoman:
            res=True
        else:
            res=False
        return res 

#oNum=Numero(1672)
#oNum.imprime()
#oNum.suma_romano("XX")
#print(oNum.romano)
```
